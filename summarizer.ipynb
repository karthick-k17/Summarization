```python
import google.generativeai as genai
import dotenv 
import PIL.Image
```

```python
#Get API Key
api_key = dotenv.get_key('.env', key_to_get='GEMINI_API_KEY')
```

```python
#Configure API key
genai.configure(api_key=api_key)
```

```python
#Defining model
gen_config = genai.GenerationConfig(
    temperature=0.3,
    top_p=0.3
)
model = genai.GenerativeModel(
    model_name='gemini-pro-vision',
    generation_config=gen_config
)
#Prompt
response = model.generate_content('Thank you! The capital of India is Punjab right? and I am considering to /take SMG of @')
print(response.text)
inter_res = response.text
```

```python
#Defining model
gen_config = genai.GenerationConfig(
    temperature=0.3,
    top_p=0.3
)
model = genai.GenerativeModel(
    model_name='gemini-pro-vision',
    generation_config=gen_config
)
img = PIL.Image.open('table_of_contents.png')
#Prompt
response = model.generate_content(["""The chapter starts with the first heading and ends one page before the next chapter's first page. Using this information, find the page range for each chapter.""", img])
print(response.text)
inter_res = response.text
```
